## 1. Prerequisites

Before installing Neurosurfer, make sure your environment meets the following requirements.

### Python version

Neurosurfer requires **Python 3.11 or later**.

```bash
python --version   # should print Python 3.11.x or higher
```

If you have an older version, install Python 3.11+ from [python.org](https://www.python.org/downloads/) or via `pyenv`.

### pip

Make sure `pip` is up to date — older `pip` versions can fail to resolve the dependency tree correctly.

```bash
pip install --upgrade pip
```

### Virtual environment (strongly recommended)

Always work inside an isolated virtual environment to avoid dependency conflicts with other projects.

**Using `venv` (built into Python):**
```bash
python -m venv .venv
source .venv/bin/activate       # Linux / macOS
.venv\Scripts\activate          # Windows
```

**Using `conda`:**
```bash
conda create -n neurosurfer python=3.11
conda activate neurosurfer
```

### GPU (optional)

A CUDA-capable GPU is only needed if you plan to run **local models** (via `[local-models]` extra).  
All cloud providers (Anthropic, OpenAI, Ollama, LM Studio) work fine on CPU-only machines.

````python
# Check your Python version before proceeding
import sys
print(f"Python {sys.version}")
assert sys.version_info >= (3, 11), "Python 3.11+ is required. Please upgrade."
````

## 2. What's in the box

Neurosurfer is a framework for building intelligent apps that blend **LLM reasoning**, **tools**, and **retrieval**.

| Module | What it provides |
|---|---|
| `neurosurfer.agents` | `AgenticLoop`, `ReactAgent`, `Agent` — three production-grade agent types |
| `neurosurfer.llm` | Providers: Anthropic, OpenAI, and any OpenAI-compatible endpoint |
| `neurosurfer.tools` | 15+ built-in tools: web search, sandboxed Python exec, file ops, HTTP, browser, memory |
| `neurosurfer.rag` | Ingest → chunk → embed → retrieve → token-aware context injection |
| `neurosurfer.graph` | DAG workflow engine + conversational Architect builder |
| `neurosurfer.server` | OpenAI-compatible FastAPI gateway with SSE streaming |
| `neurosurfer.cache` | LLM response cache + embedding cache |
| `neurosurfer.memory` | Long-term memory: distillation + semantic retrieval |

Everything in the table above is included in the **base install** — no extras needed.

## 6. Setting Up API Keys

Neurosurfer reads API keys from **environment variables**. No keys are hard-coded anywhere.

### Which key do you need?

| Provider | Environment variable | Where to get it |
|---|---|---|
| Anthropic (Claude) | `ANTHROPIC_API_KEY` | [console.anthropic.com](https://console.anthropic.com/) |
| OpenAI (GPT-4o, etc.) | `OPENAI_API_KEY` | [platform.openai.com](https://platform.openai.com/api-keys) |
| SerpAPI (Google search) | `SERPAPI_API_KEY` | [serpapi.com](https://serpapi.com/) — optional; DuckDuckGo works without one |
| Ollama / LM Studio / vLLM | no key needed | These run locally; point the provider at `http://localhost:<port>` |

### Option A — Set in your shell (recommended for local development)

Add these lines to your shell profile (`~/.bashrc`, `~/.zshrc`, etc.) so they persist across sessions:

```bash
export ANTHROPIC_API_KEY="sk-ant-..."
export OPENAI_API_KEY="sk-..."
```

Then reload:
```bash
source ~/.bashrc   # or ~/.zshrc
```

1. Create a `.env` file at your project root (never commit this file — add it to `.gitignore`):


### Using a local provider (no API key needed)

If you're running Ollama, LM Studio, or any other OpenAI-compatible local server, no key is required.  
You only need to point the provider at the right URL.

**`context_window` behaviour:**

| Model type | `context_window` |
|---|---|
| Known OpenAI frontier models (`gpt-4o`, `o3`, `o4-mini`, …) | Auto-resolved — you don't need to pass it |
| Local / custom models (LM Studio, Ollama, vLLM, …) | **Required** — pass it explicitly |

If you omit `context_window` for an unknown model, you'll get a clear `ValueError` telling you what to pass.

```python
from neurosurfer.llm.providers.openai import OpenAICompatProvider

# ── Local models: context_window required ──────────────────────────────────

# LM Studio
provider = OpenAICompatProvider(
    model="google/gemma-4-12b-qat",
    base_url="http://localhost:1234/v1",
    api_key="lm-studio",     # any non-empty string
    context_window=32_768,   # match your model's actual context size
)

# Ollama
provider = OpenAICompatProvider(
    model="llama3.1:8b",
    base_url="http://localhost:11434/v1",
    api_key="ollama",
    context_window=131_072,
)

# ── OpenAI frontier models: context_window auto-resolved ───────────────────

provider = OpenAICompatProvider(
    model="gpt-4o-mini",
    base_url="https://api.openai.com/v1",
    api_key=os.environ["OPENAI_API_KEY"],
    # context_window omitted — resolved to 128,000 automatically
)
```

> `OpenAICompatProvider` works with **any server that implements the OpenAI `/v1/chat/completions` API** — Ollama, LM Studio, vLLM, llama.cpp, the Neurosurfer gateway, and OpenAI itself.

## 8. Quick Smoke Test

This is a minimal end-to-end check: initialize a provider and make one API call.  
It confirms your API key is valid and the network connection works.

Run **one** of the blocks below depending on which provider you have set up.

> **Note:** This is a real API call — it costs a tiny fraction of a cent.  
> Skip this section if you prefer to test in a later tutorial.

````python
# Smoke test — Anthropic (Claude)
# Requires ANTHROPIC_API_KEY to be set.

import asyncio
from neurosurfer.llm.providers.anthropic import AnthropicProvider
from neurosurfer.llm.types import Message, GenerationConfig

provider = AnthropicProvider(model="claude-haiku-4-5-20251001")   # fast + cheap

messages = [Message.user_text("Reply with exactly three words: installation was successful.")]
config   = GenerationConfig(max_tokens=32, temperature=0.0)

async def smoke_test():
    response = await provider.complete(messages=messages, config=config)
    print("Anthropic response:", response.text)

await smoke_test()
````

````python
# Smoke test — local provider (Ollama / LM Studio / any OpenAI-compatible server)
# No API key required. Adjust base_url and model to match your server.
#
# context_window is required for OpenAICompatProvider — local servers don't
# advertise their context size, so you must pass it explicitly.

import asyncio
from neurosurfer.llm.providers.openai import OpenAICompatProvider
from neurosurfer.llm.types import Message, GenerationConfig

provider = OpenAICompatProvider(
    model="google/gemma-4-12b-qat",
    base_url="http://localhost:1234/v1",   # LM Studio; Ollama uses :11434
    api_key="lmstudio",                   # any non-empty string for local servers
    context_window=32_768,
)

messages = [Message.user_text("Reply with exactly three words: installation was successful.")]
config   = GenerationConfig(max_tokens=32, temperature=0.0)

async def smoke_test():
    response = await provider.complete(messages=messages, config=config)
    print("Local model response:", response.text)

await smoke_test()
````

### Troubleshooting

| Problem | Fix |
|---|---|
| `ModuleNotFoundError: neurosurfer` | Confirm you activated the right virtual environment before installing |
| `AuthenticationError` | Check your API key is set correctly in the environment (`echo $ANTHROPIC_API_KEY`) |
| `ImportError: No module named 'chromadb'` | Install the `rag` extra: `pip install "neurosurfer[rag]"` |
| `ImportError: No module named 'fastapi'` | Install the `serve` extra: `pip install "neurosurfer[serve]"` |
| `neurosurfer: command not found` | Your venv's `bin/` is not on PATH; run `source .venv/bin/activate` |
| Slow installs on Colab | Use `--quiet` flag: `pip install -q neurosurfer` |

If you hit something not listed here, open an issue at [github.com/NaumanHSA/neurosurfer/issues](https://github.com/NaumanHSA/neurosurfer/issues).
